```python
### Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
mapping = {}
cat_cols = ['hotel','arrival_date_month','meal','country',
            'market_segment','distribution_channel' ,
            'reserved_room_type','phone-number','credit_card']
for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])
    mapping[col] = {label:code for label,code in zip(label_encoder.classes_,label_encoder.transform(label_encoder.classes_)) }
```
